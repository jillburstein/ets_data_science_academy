```python
#create rectangle class and compute area; *NOT SURE WHAT IS MEANT BY 'SHIFT' IN THE HW; 
#COULDN'T FIGURE OUT HOW TO CREATE A SHAPE CLASS AND THEN HAVE IT INTERACT WITH THE RECTANGLE/CIRCLE
#I FOUND AN EXAMPLE, BUT TBH, DID NOT REALLY UNDERSTAND HOW IT WORKED: https://overiq.com/python-101/inheritance-and-polymorphism-in-python/#:~:text=In%20lines%203%2D19%2C%20we,setter%20methods%20for%20those%20attributes.

import math

class Rectangle:

    def __init__ (self, length, width):
        self.length = length
        self.width = width

    def get_length(self): #return and use as arg1
        return self.length
    
    def get_width(self): #return and use as arg2
        return self.width

    def get_area(self): #compute area
        return self.length * self.width
```

```python
rectangle = Rectangle(5,10)

print("Area of this rectangle is:", rectangle.get_area())
```

```python
#create circle class and compute area
class Circle:

    def __init__(self, radius):
        self.radius = radius

    def get_radius(self):
        return self.radius

    def get_area(self):
        return math.pi * self.radius ** 2

```

```python
circle = Circle(5)

print("Area of this circle is:", circle.get_area())
```
