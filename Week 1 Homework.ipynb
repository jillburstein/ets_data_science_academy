```python
#18. Consider a 16x16 array, how to get the block-sum (block size is 4x4)?
#UPDATED

def contiguous_4x4_blocks(a):
    # a.shape is a tuple, and we can extract its values into two variables m, n with one line instead of
    # m = a.shape[0]
    # n = a.shape[1]
    m, n = a.shape
    #print("m=",m,"n=",n)
    #print(range(m-4),range(n-4) )
    return [a[i:i+4, j:j+4] for i in range(m-4) for j in range(n-4)]

"""Let's test this but not on a random matrix so it's easy for us to verify this by eye."""

a = np.arange(256).reshape(16,16)
print(a)
for block in contiguous_4x4_blocks(a):
    print("block=\n",block)
    print ('Overall sum of block is %s\n'%np.sum(block))
```
