```python
import pandas as pd
import xml.etree.ElementTree as ET
tree = ET.parse('hw_w2_xml_parsing.xml')
root = tree.getroot()
```

```python
xml_csv = open('xml.csv', "w")

create_header = "Event,Text"
xml_csv.write(create_header)
xml_csv.write("\n")

print("ROOT:",root.tag[55:],"==>")
for i in range(len(root)): #get root length to iterate
    for chd in root[i]: 
        print(root[i].tag[55:],"==>",chd.tag[55:])
        s1 = str(chd.tag[55:]).strip()
        s2 = str(chd.text).strip()
        s3 = str(s1 + "," + s2 + "\n")
        xml_csv.write(s3)
        
        for j in range(len(chd)): #get child length to iterate within children
            print("==>",chd[j].tag[55:],"==>")  
            s4 = str(chd[j].tag[55:]).strip()
            #s5 = str(chd.text).strip()
            s6 = str(chd[j].text).strip()
            if "," in s6: #check if "," is in text
                s6 = s6.replace(",","+")
            s7 = str(s4 + "," + s6 + "\n")
            xml_csv.write(s7)
          
            for k in range(len(chd[j])):
                print("==>",chd[j][k].tag[55:], "==>", chd[j][k].text)
                s8 = str(chd[j][k].tag[55:]).strip()
                s9 = str(chd[j][k].text).strip()
                s10= str(s8 + "," + s9 + "\n")
                xml_csv.write(s10)
                
```
